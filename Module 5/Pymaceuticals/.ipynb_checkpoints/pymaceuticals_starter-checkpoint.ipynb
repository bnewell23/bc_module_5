## Analysis of Drug Regimens on Tumor Volume

In this analysis, we investigated the effectiveness of various drug regimens in reducing tumor volumes in a study. Four drug regimens were examined: Capomulin, Ramicane, Infubinol, and Ceftamin.

### 1. Tumor Volume Distribution:

**Capomulin and Ramicane:**
- Both regimens showed similar results with lower median tumor volumes compared to other regimens.
- No outliers were detected, suggesting consistent efficacy within the treatment group.

**Infubinol:**
- Although Infubinol exhibited a wider interquartile range (IQR) compared to Capomulin and Ramicane, it had one outlier with a notably lower tumor volume.
- Despite the outlier, Infubinol's median tumor volume remained relatively high.

**Ceftamin:**
- Ceftamin displayed a wider spread of tumor volumes with no outliers detected.
- The median tumor volume for Ceftamin was higher compared to Capomulin and Ramicane but lower than Infubinol.

### 2. Correlation Between Drug Regimens and Tumor Volume:

A strong positive correlation coefficient of approximately 0.842 was observed between drug regimen and tumor volume. This suggests that as the effectiveness of the drug regimen increases, tumor volume tends to decrease.

### 3. Summary Statistics:

- Mean tumor volume was lowest for Capomulin and Ramicane, indicating their effectiveness in reducing tumor size.
- Ceftamin and Infubinol exhibited higher mean tumor volumes, suggesting less efficacy compared to Capomulin and Ramicane.
- The standard error of the mean (SEM) was relatively low for all drug regimens, indicating the reliability of the mean estimates.

### Conclusion:

In conclusion, Capomulin and Ramicane emerged as the most effective drug regimens in reducing tumor volumes, as evidenced by their lower median and mean tumor volumes and absence of outliers. While Infubinol showed promising results despite one outlier, Ceftamin exhibited less efficacy compared to the other regimens. Further studies may be warranted to explore potential factors contributing to the variation in treatment outcomes across different drug regimens.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame using left merge
merged_data = pd.merge(study_results, mouse_metadata, on='Mouse ID', how='left')

# Display the data table for preview
display(merged_data.head())
```

```python
# Checking the number of mice.
num_mice = merged_data['Mouse ID'].nunique()
print("Number of Mice:", num_mice)
```

```python
# Identify duplicate mice by Mouse ID and Timepoint
duplicate_mice = merged_data[merged_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]

# Get unique IDs of duplicate mice
duplicate_ids = duplicate_mice['Mouse ID'].unique()

print("Duplicate mice IDs:", duplicate_ids)
```

```python
# Filter the merged data for the duplicate mouse ID 'g989'
duplicate_data_g989 = merged_data[merged_data['Mouse ID'] == 'g989']

display(duplicate_data_g989)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID
cleaned_data = merged_data.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='first')

# Display the cleaned DataFrame
display(cleaned_data.head())
```

```python
# Checking the number of mice in the clean DataFrame
num_mice_cleaned = cleaned_data['Mouse ID'].nunique()
print("Number of Mice in the Clean DataFrame:", num_mice_cleaned)
```

## Summary Statistics

```python
# Group by 'Drug Regimen' and calculate the desired statistics
summary_stats = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Rename the columns
summary_stats = summary_stats.rename(columns={'mean': 'Mean Tumor Volume',
                                              'median': 'Median Tumor Volume',
                                              'var': 'Tumor Volume Variance',
                                              'std': 'Tumor Volume Std. Dev.',
                                              'sem': 'Tumor Volume Std. Err.'})

# Display the summary DataFrame
display(summary_stats)
```

```python
# Using the aggregation method to produce the summary statistics
summary_stats_advanced = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Display the summary DataFrame
display(summary_stats_advanced)
```

## Quartiles, Outliers and Boxplots

```python
# Step 1: Get the last (greatest) timepoint for each mouse
last_timepoint = cleaned_data.groupby('Mouse ID')['Timepoint'].max()

# Step 2: Merge this group DataFrame with the original DataFrame to get the tumor volume at the last timepoint
merged_last_timepoint = pd.merge(last_timepoint, cleaned_data, on=['Mouse ID', 'Timepoint'], how='inner')

# Filter for the specified treatment regimens: Capomulin, Ramicane, Infubinol, and Ceftamin
selected_treatments = merged_last_timepoint[merged_last_timepoint['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]

# Display the DataFrame showing the final tumor volume for each mouse across the selected treatment regimens
display(selected_treatments[['Mouse ID', 'Drug Regimen', 'Timepoint', 'Tumor Volume (mm3)']])
```

```python
# Put treatments into a list for for loop
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create an empty list to fill with tumor volume data
tumor_volume_data = []

# Loop through each treatment regimen
for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volumes = selected_treatments[selected_treatments['Drug Regimen'] == treatment]['Tumor Volume (mm3)']
    
    # Append tumor volumes to the list
    tumor_volume_data.append(tumor_volumes)


# Calculate the IQR and quantitatively determine if there are any potential outliers
for i, treatment in enumerate(treatments):
    # Calculate quartiles, IQR, and bounds
    quartiles = tumor_volume_data[i].quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    
    # Print results
    print(f"--------{treatment}--------")
    print(f"IQR: {iqr}")
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    
    # Check for outliers
    outliers = tumor_volume_data[i][(tumor_volume_data[i] < lower_bound) | (tumor_volume_data[i] > upper_bound)]
    if not outliers.empty:
        print("Outliers:")
        print(outliers)
    else:
        print("No outliers")
    print("\n")
```

```python
# Importing necessary library
import matplotlib.pyplot as plt

# Generate a box plot for each treatment group
plt.figure(figsize=(10, 6))
plt.boxplot(tumor_volume_data, labels=treatments, flierprops=dict(marker='o', markersize=8, markerfacecolor='red', linestyle='none'))

# Adding labels and title
plt.title('Distribution of Tumor Volume for Each Treatment Group')
plt.xlabel('Treatment Group')
plt.ylabel('Tumor Volume (mm3)')

# Display the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Line and Scatter Plots

```python
# Select a single mouse ID treated with Capomulin
mouse_id = 's185'

# Filter the data for the selected mouse ID and Capomulin treatment regimen
mouse_data_capomulin = cleaned_data[(cleaned_data['Mouse ID'] == mouse_id) & (cleaned_data['Drug Regimen'] == 'Capomulin')]

# Plot tumor volume vs. time point
plt.figure(figsize=(10, 6))
plt.plot(mouse_data_capomulin['Timepoint'], mouse_data_capomulin['Tumor Volume (mm3)'], marker='o', color='skyblue')

# Adding labels and title
plt.title(f'Tumor Volume vs. Time Point for Mouse {mouse_id} (Capomulin)')
plt.xlabel('Time Point')
plt.ylabel('Tumor Volume (mm3)')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Group the data by Mouse ID to calculate the average tumor volume for each mouse
average_tumor_volume = cleaned_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Step 2: Merge the calculated average tumor volume with the mouse metadata DataFrame to include mouse weight
average_tumor_volume_df = pd.merge(average_tumor_volume, mouse_metadata, on='Mouse ID', how='left')

# Step 3: Filter the data to include only mice treated with the Capomulin regimen
capomulin_data = average_tumor_volume_df[average_tumor_volume_df['Drug Regimen'] == 'Capomulin']

# Step 4: Plot mouse weight vs. the average tumor volume
plt.figure(figsize=(10, 6))
plt.scatter(capomulin_data['Weight (g)'], capomulin_data['Tumor Volume (mm3)'], color='skyblue')

# Adding labels and title
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient
correlation_coefficient, _ = st.pearsonr(capomulin_data['Weight (g)'], capomulin_data['Tumor Volume (mm3)'])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = st.linregress(capomulin_data['Weight (g)'], capomulin_data['Tumor Volume (mm3)'])

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(capomulin_data['Weight (g)'], capomulin_data['Tumor Volume (mm3)'], color='skyblue', label='Data')

# Plot the linear regression line
plt.plot(capomulin_data['Weight (g)'], slope * capomulin_data['Weight (g)'] + intercept, color='red', label='Linear Regression')

# Adding labels and title
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Adding legend
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()

# Print the correlation coefficient and linear regression results
print("Correlation coefficient:", correlation_coefficient)
```
